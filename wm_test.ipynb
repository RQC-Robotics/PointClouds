```python
import sys;
sys.path.insert(0, '~/Documents/RL/RLTools/') # insert_path
from rltools.algos.dreamer.dreamer import Dreamer
from rltools.env.wrappers.dm_control import dmWrapper
from rltools.env.wrappers.gym import FrameSkip
from dm_control import suite
```

```python
env = FrameSkip(dmWrapper(suite.load('cartpole', 'balance')), 6)
n_actions = env.action_space.shape[0]

agent = Dreamer(obs_size=len(env.reset()),
                action_size=env.action_space.shape[0],
                hidden_size=100,
                latent_size=30)
agent.compile(device='cuda')
```

```python
agent.learn(env)
```
